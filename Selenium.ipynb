# Scraper: Grab Manila (Selenium)

```python
from selenium import webdriver
import time
import pandas as pd
import json

url = 'https://food.grab.com/ph/en/restaurants'
```

```python
wd = webdriver.Chrome()
wd.get(url)
```

```python
for i in range(5):
    
    try:    
        wd.find_element_by_xpath("//button[@class='ant-btn ant-btn-block']").click()
        time.sleep(3)
    except:
        pass
        time.sleep(3)    
```

```python
page = wd.find_elements_by_xpath("//script[@id='__NEXT_DATA__'][@type='application/json']")
```

### Created two files restaurants.json & restaurants2.json by pasting the above string on "jsonformatter.curiousconcept.com"

```python
f = open('restaurants.json')
data = json.load(f)

name = []
latitude = []
longitude = []

for key in data:
    name.append(data[key]['name'])
    latitude.append(data[key]['latitude'])
    longitude.append(data[key]['longitude'])

df = pd.DataFrame(list(zip(name, latitude, longitude)), columns=['Name', 'Latitude', 'Longitude'])
```

```python
f = open('restaurants2.json')
data = json.load(f)

name = []
latitude = []
longitude = []

for key in data:
    name.append(data[key]['name'])
    latitude.append(data[key]['latitude'])
    longitude.append(data[key]['longitude'])

df2 = pd.DataFrame(list(zip(name, latitude, longitude)), columns=['Name', 'Latitude', 'Longitude'])
```

```python
df = pd.concat([df, df2])
df.drop_duplicates(inplace=True, ignore_index=True)
```
