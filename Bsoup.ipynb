# Scraper: Grab Manila (Beautiful Soup) 

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request
from requests import get
from bs2json import bs2json
import json
```

```python
headers = ({'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_5_2) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.1.2 Safari/605.1.15'})
```

```python
url = 'https://food.grab.com/ph/en/restaurants'

request = Request(url, headers=headers)
response = urlopen(request)

html = response.read()
html_soup = BeautifulSoup(html ,'html.parser')
```

```python
print(html_soup.prettify())
```

```python
rest_container = html_soup.find('script', id='__NEXT_DATA__')
```

```python
converter = bs2json()
py_dict = converter.convert(rest_container)
```
